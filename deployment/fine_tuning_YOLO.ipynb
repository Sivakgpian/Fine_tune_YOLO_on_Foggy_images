```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")
model.train(
    data="data_annotated/data.yaml",
    epochs=10,
    imgsz=640,
    batch=1,
    name="foggy_yolov8s_run1",
    patience=20
)
```

```python
# Load the trained model
model = YOLO("yolov8s.pt") 

metrics = model.val(
    data="data_annotated/data.yaml",
    imgsz=640,
    split="val"  
)

print(metrics)
```

```python
from ultralytics import YOLO

# Load your trained model
model = YOLO("runs/detect/your_foggy_yolov8s_run1/weights/yolov8s.pt")

# Export to ONNX format
model.export(format="onnx", imgsz=640, half=True)
```

```python
import onnx

onnx_model_path = 'runs/detect/your_foggy_yolov8s_run1/weights/yolov8s.onnx' 

onnx_model = onnx.load(onnx_model_path)

try:
    onnx.checker.check_model(onnx_model)
    print("ONNX model check passed!")
except onnx.checker.ValidationError as e:
    print(f"ONNX model check failed: {e}")

print("\nModel Inputs:")
for input_node in onnx_model.graph.input:
    print(f"  Name: {input_node.name}, Shape: {[dim.dim_value for dim in input_node.type.tensor_type.shape.dim]}")

print("\nModel Outputs:")
for output_node in onnx_model.graph.output:
    print(f"  Name: {output_node.name}, Shape: {[dim.dim_value for dim in output_node.type.tensor_type.shape.dim]}")
```
